```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the dataset
file_path = r"F:\cyber security project\mahesh\data\Train_data.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print("Initial DataFrame:")
print(df.head())

# Display basic info
print("\nDataFrame Info:")
print(df.info())

# Handle missing values if any (here we drop rows with missing values)
df.dropna(inplace=True)
print("\nDataFrame after dropping missing values:")
print(df.info())

# Convert categorical features to numerical values
# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
print("\nCategorical columns:")
print(categorical_cols)

# Apply Label Encoding to categorical columns
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

print("\nDataFrame after Label Encoding:")
print(df.head())

# Calculate correlation matrix
correlation_matrix = df.corr()

# Identify highly correlated features (threshold = 0.9)
threshold = 0.9
highly_correlated_features = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname = correlation_matrix.columns[i]
            highly_correlated_features.add(colname)

# Drop highly correlated features
df.drop(columns=highly_correlated_features, inplace=True)
print("\nDataFrame after dropping highly correlated features:")
print(df.head())

# Split the data into features and target
X = df.drop('class', axis=1)
y = df['class']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

print("\nShape of Training and Testing Sets:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```
